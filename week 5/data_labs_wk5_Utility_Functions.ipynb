```python
import sqlite3

import sqlite3

def show_create_table(db_file, table_name):

    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    query = """
    SELECT sql
    FROM sqlite_master
    WHERE type='table' AND name=?;
    """

    cursor.execute(query, (table_name,))
    result = cursor.fetchone()

    if result:
        print(f"\nCREATE statement for table '{table_name}':\n")
        print(result[0])
    else:
        print("Table not found.")

    conn.close()

def inspect_database(db_file):

    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    cursor.execute("""
        SELECT name
        FROM sqlite_master
        WHERE type='table';
    """)

    tables = cursor.fetchall()

    for table in tables:

        table_name = table[0]
        print(f"\nTable: {table_name}")
        print("-" * 40)

        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = cursor.fetchall()

        for col in columns:
            print(f"Column: {col[1]} | Type: {col[2]}")

    conn.close()

def show_all_create_statements(db_file):

    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    cursor.execute("""
    SELECT name, sql
    FROM sqlite_master
    WHERE type='table';
    """)

    tables = cursor.fetchall()

    for table in tables:
        print("\nTable:", table[0])
        print(table[1])
        print("-" * 50)

    conn.close()

#inspect_database("example.db")  #use this for a rapid inspection of your tables
# Example usage    
#show_create_table("example.db", "passenger") #Use this for a detailed inspection of your tables incdluing how it was created
show_all_create_statements("lusitania_many_to_many.db")
```
