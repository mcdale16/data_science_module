```python
import sqlite3
import pandas as pd
```

```python
csv_file = "LusitaniaManifest.csv"
```

```python
db_file = "example.db"
```

```python
table_name = "ship_data"
```

```python
def connect_to_database(db_file):
    con = sqlite3.connect(db_file)
    print(con)
    return con

def read_data_into_database(db_base, csv_file, table_name):
    df = pd.read_csv(csv_file)
    df.to_sql(table_name, db_base, if_exists = "replace", index = False)

def read_data_from_database(cursor, table_name):
    cursor.execute("SELECT * FROM " + table_name)
    rows = cursor.fetchall()
    for row in rows:
        print(rows)
```

```python
db_base = connect_to_database(db_file)
read_data_into_database(db_base, csv_file, table_name)
cursor = db_base.cursor()
read_data_from_database(cursor, table_name)
```
